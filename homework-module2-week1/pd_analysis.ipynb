```python
import pandas as pd
import numpy as np
# Load the CSV file
df = pd.read_csv('./advertising.csv')

# Convert the DataFrame to a numpy array
data = df.to_numpy()
data.shape
```

```python
A = np.average(data[:,3])
new_arr = np.empty(data.shape, dtype=object)
new_arr[data>A] = 'Good'
new_arr[data==A] = 'Average'
new_arr[data<A] = 'Bad'

new_arr.reshape(-1,)[7:10]
new_arr
```

```python
less_than_avg_data = data[data[:,3] < np.average(data[:,3])]
sorted_data = less_than_avg_data[less_than_avg_data[:, 3].argsort()]
A = sorted_data[-1,3]
new_arr[data>A] = 'Good'
new_arr[data==A] = 'Average'
new_arr[data<A] = 'Bad'
new_arr.reshape(-1,)[7:10]
```
