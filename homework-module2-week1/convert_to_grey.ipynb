```python
import matplotlib.image as mpimg
import numpy as np

img = mpimg.imread('./dog.jpeg')

min_val, max_val = np.min(img, axis=2), np.max(img, axis=2)
min_val.shape, max_val.shape
```

```python
ligness_image = (min_val + max_val) / 2
ligness_image[0,0]
```

```python
avg_image = np.average(img, axis=2)
avg_image[0,0]
```

```python
lumi_img = np.zeros((img.shape[0], img.shape[1]), dtype=np.int32) 
lumi_img = 0.21*(img[:,:,0].astype(np.uint16)) + 0.72*(img[:,:,1].astype(np.uint16)) + 0.07*(img[:,:,2].astype(np.uint16))
lumi_img[0][0]
```

```python
import matplotlib.pyplot as plt

plt.imshow(ligness_image, cmap='gray')
plt.axis('off')  # Hide axis
plt.show()
```

```python
plt.imshow(avg_image, cmap='gray')
plt.axis('off')  # Hide axis
plt.show()
```

```python
plt.imshow(lumi_img, cmap='gray')
plt.axis('off')  # Hide axis
plt.show()
```
