```python
import numpy as np
import cv2 

green_background = cv2.imread('./images/GreenBackground.png',1)
green_background = cv2.cvtColor(green_background, cv2.COLOR_BGR2RGB)
green_background = cv2.resize(green_background, (678 , 381))

new_background = cv2.imread('./images/NewBackground.jpg',1)
new_background = cv2.cvtColor(new_background, cv2.COLOR_BGR2RGB)
new_background = cv2.resize(new_background, (678 , 381))

object_image = cv2.imread('./images/Object.png',1)
object_image = cv2.cvtColor(object_image, cv2.COLOR_BGR2RGB)
object_image = cv2.resize(object_image, (678 , 381))

green_background.shape, new_background.shape, object_image.shape
```

```python
import  matplotlib.pyplot as plt

def compute_differernce(img1, img2):

    result = img1 - img2

    return result

diff_img = compute_differernce(object_image, green_background)
plt.imshow(diff_img)
plt.show()
```

```python
def compute_binary_mask(img):

    binary_image = np.where(img == 0, 0, 255)

    return binary_image

binary_image = compute_binary_mask(diff_img)
plt.imshow(binary_image)
plt.show()
```

```python
def replace_background(green_background, new_background, object_image):

    diff_image = compute_differernce(green_background, object_image) 
    binary_image = compute_binary_mask(diff_image)

    image_with_new_object = np.where(binary_image == 255, object_image, new_background)

    return image_with_new_object

image_with_new_object = replace_background(green_background, new_background, object_image)
plt.imshow(image_with_new_object)
plt.show()
```
